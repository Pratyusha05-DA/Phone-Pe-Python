### AGGREGATED INSURANCE 

```python
import os
import pandas as pd
import json

# Define the base directory
base_dir = r"C:\Users\hi\Downloads\Phone Pe project\pulse-master\data\aggregated\insurance\country\india\state"
# Initialize an empty list to store the data
data = []

# Traverse the directory structure
for state_folder in os.listdir(base_dir):
    # Replace '-' with ' ' and '&' with 'and' in the state name, and capitalize the first letter
    state_name = state_folder.replace('-', ' ').replace('&', 'and').title()
    
    state_path = os.path.join(base_dir, state_folder)
    if os.path.isdir(state_path):
        for year_folder in os.listdir(state_path):
            year_path = os.path.join(state_path, year_folder)
            if os.path.isdir(year_path):
                for quarter_file in os.listdir(year_path):
                    if quarter_file.endswith('.json'):
                        quarter = quarter_file.split('.')[0]
                        json_path = os.path.join(year_path, quarter_file)
                        with open(json_path, 'r') as file:
                            json_data = json.load(file)
                            # Ensure 'data' and 'transactionData' keys are present
                            if 'data' in json_data and 'transactionData' in json_data['data']:
                                transaction_data_list = json_data['data']['transactionData']
                                for data_item in transaction_data_list:
                                    transaction_type = data_item['name']
                                    for payment_instrument in data_item['paymentInstruments']:
                                        count = payment_instrument['count']
                                        amount = payment_instrument['amount']
                                        data.append([state_name, year_folder, quarter, transaction_type, count, amount])

# Convert the list to a DataFrame
df1 = pd.DataFrame(data, columns=[
    'State', 'Year', 'Quarter', 'Transaction_Type', 'Transaction_Count', 'Transaction_Amount'
])

# Save DataFrame to CSV
df1.to_csv('Aggregate Insurance.csv', index=False)
```

### * AGGREGATED USER ( 2018 - 2022 )

```python
import os
import pandas as pd
import json

# Define the base directory
base_dir = r"C:\Users\hi\Downloads\Phone Pe project\pulse-master\data\aggregated\user\country\india\state"

# Initialize an empty list to store the data
data = []

# Traverse the directory structure
for state_folder in os.listdir(base_dir):
    # Replace '-' with ' ' and '&' with 'and' in the state name, and capitalize the first letter
    state_name = state_folder.replace('-', ' ').replace('&', 'and').title()
       
    state_path = os.path.join(base_dir, state_folder)
    if os.path.isdir(state_path):
        for year_folder in os.listdir(state_path):
            year_path = os.path.join(state_path, year_folder)
            if os.path.isdir(year_path):
                for quarter_file in os.listdir(year_path):
                    if quarter_file.endswith('.json'):
                        quarter = quarter_file.split('.')[0]
                        json_path = os.path.join(year_path, quarter_file)
                        with open(json_path, 'r') as file:
                            json_data = json.load(file)
                            # Ensure 'data' and 'usersByDevice' keys are present and not None
                            if ('data' in json_data and 
                                'usersByDevice' in json_data['data'] and 
                                json_data['data']['usersByDevice'] is not None):
                                
                                users_by_device = json_data['data']['usersByDevice']
                                for device in users_by_device:
                                    brand = device['brand']
                                    count = device['count']
                                    percentage = device['percentage']
                                    data.append([state_name, year_folder, quarter, brand, count, percentage])

# Convert the list to a DataFrame
df3 = pd.DataFrame(data, columns=['State', 'Year', 'Quarter', 'Brand', 'Transaction_count', 'Percentage'])

# Save DataFrame to CSV
df3.to_csv('Aggregate User.csv', index=False)
```
